```python
import torch
x = torch.arange(24).reshape(2, 3, 4)
print(x.shape)
print(x)
```

```python
y = x.transpose(1, 2)
print(y.shape)
print(y)
```

```python
z = x.transpose(0, 1)
print(z.shape)
print(z)
```

```python
import torch

bs, seq_len, vocab = 2, 4, 10
x = torch.randn(bs, seq_len, vocab)
print(x.shape)

indices = torch.randint(0, vocab, (bs, seq_len, 3))

y = torch.gather(x, 2, indices)
print(y.shape)
```
