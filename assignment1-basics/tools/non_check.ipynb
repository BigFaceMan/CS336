```python
vocab = {}
for i in range(256):
    vocab[bytes([i])] = i
special_tokens = ["你好"]
for idx, sp_tk in enumerate(special_tokens):
    vocab[sp_tk.encode("utf-8")] = 256 + idx
for key, val in vocab.items():
    if val > 255:
        print(key.decode("utf-8"), val)
```
